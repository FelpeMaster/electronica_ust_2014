# Diodos Zener

En el siguiente circuito $V_z = 20V, \ v_{in} = 60 \sin{\omega t}, \ V_{\gamma} = 0,7V$

```python
from IPython.core.display import HTML 
HTML('<img src="zener1.svg" width=500 height=100/>') 
```

Sí analizamos aplicando la Ley de Kirchhoff de voltaje durante la primera mitad del período de la onda de entrada se cumple que  $v_{in}(t)>0$.

\begin{eqnarray}
-v_{in} + iR + V_{Z} + V_{\gamma} = 0\\
-60 \sin{\left(\omega t\right)} V + iR + 20V + 0,7V = 0 \\
i = \frac{60 \sin{\left(\omega t\right)} V - 20,7V}{R}
\end{eqnarray}

Sí $i>0A$ entonces el circuito debe ser capaz de hacer que por el diodo zener se polarice a $20V$ y el diodo rectificador esté directamente polarizado. De ese modo se puede obtener la siguiente relación.

\begin{eqnarray}
i = \frac{60 \sin{\left(\omega t\right)} V - 20,7V}{R}>0\\
\Rightarrow 60 \sin{\left(\omega t\right)} V - 20,7V >0 \\
60\sin{\left(\omega t\right)} V > 20,7V\\
\therefore v_{out}(t) = 20V
\end{eqnarray}

Cuando el voltaje de entrada es menor a $20,7V$ los diodos estarán polarizados en forma inversa, en particular el diodo rectificador se polarizará en forma inversa, por lo tanto el voltaje de salida en el diodo Zener será igual a cero, esto se puede observar en la siguiente figura.

```python
%pylab inline
import numpy as np
import pylab as pl
t = np.arange(0,20,0.001)
w = np.pi*2*0.1
vin = 60*np.sin(w*t)
vout = np.zeros(len(t))
for i in range(len(t)):
    if vin[i]>20.7:
        vout[i] = 20
pl.figure(figsize=(15, 6), dpi=1000)
pl.plot(t,vin,t,vout)
ax = pl.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0.))
pl.ylabel('Voltaje',fontsize = 20)
pl.xlabel('tiempo',fontsize = 20)
pl.legend(['$v_{in}(t)$','$v_{out}(t)$'], loc=1,fontsize = 20)
pl.grid()
pl.show()
```
