# Señales

## Señales de corriente continua.

Se caracterizan por tener un valor constante a lo largo del tiempo

```python
%pylab inline
import numpy as np
import pylab as pl
t = np.arange(0,10,0.00001)
V = 5*np.ones(len(t))
pl.plot(t,V)
pl.show()
```

## Señales de corriente alterna

Se caracterizan por tener varios valores a lo largo del tiempo

```python
t = np.arange(0,10,0.00001)
V = 5*np.sin(t)
pl.plot(t,V)
pl.show()
```

existe una equivalencia.

$\pi \ rad= 180^o$

por lo tanto si tienen una angulo $\alpha$ en radianes se puede hacer una equivalencia a grados sexagesimales $\beta$ por medio del facto de conversión. De modo que
\begin{equation}
\beta = \alpha \times \frac{180^o}{\pi \ rad} 
\end{equation}

Lo inverso también es válido
\begin{equation}
\alpha = \beta \times \frac{\pi \ rad}{180^o} 
\end{equation}

ejemplo si $\beta = \frac{\pi}{6} \rightarrow \ \alpha = \frac{\pi}{6} \times \frac{180^o}{\pi \ rad} $


$\alpha = 30^o$

ejemplo si $\alpha = 45^o \rightarrow \ \beta = 45^o \times \frac{\pi \ rad}{180^o} $

$\beta = \frac{\pi}{4}$

# señal sinuidal

$V = A \times  sin\left(\omega t\right), $ donde $A$ es la amplitud de la señal y, $\omega :$ es la valocidad angular (se mide en rad/seg)

```python
w = 1
A = 10
t = np.arange(0,10,0.0001)
v = A*np.sin(w*t)
v2 = A*np.sin(0.5*w*t)
pl.figure(figsize=(18, 8), dpi=1000)
pl.plot(t,v,t,0.5*v,t,v2, t, 1.5*v)
pl.legend(['A=10, $\omega = 1$','A=5, $\omega = 1$','A=10, $\omega = 0,5$','A=15, $\omega = 1$'],bbox_to_anchor=(0.1, 0.2), loc=2, borderaxespad=0., fontsize = 20)
pl.xlabel('tiempo, en segundos',fontsize = 20)
pl.ylabel('f(t), en Volts', fontsize = 20)
pl.show()
```
