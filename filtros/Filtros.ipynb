Una forma de hacer filtros es trabajando con circuitos eléctricos con elementos pasivos como condensadores, bobinas y resistencias. Para ello recordaremos sus expresiones más comunes.

\begin{eqnarray}
v_C = \frac{1}{C}\int_{0}^t i_C dt + v_{C0}\\
i_C = \frac{dv_C}{dt}\\
v_L =\frac{di_L}{dt}\\
i_L = \frac{1}{L}\int_{0}^t v_L dt + i_{L0}\\
v_R = iR\\
i_R = \frac{v}{R}
\end{eqnarray}

Al momento de armar circuitos eléctricos complejos hace dificilmente manejable establecer las ecuaciones. Motivo por el cual se recurre a hacer un cambio. Se introdice la impedancia, este es un parámetro electrico semejante a la resistencia electrica con la diferencia de que se trabaja en el domino de la frecuencia con numeros complejos.

## Solución.

si reemplazamos al condensador por la impedancia se tiene que $Z_C = \frac{1}{j \omega C}$ y $Z_R = R$ y aplicando la Ley de LKV. Se puede obtener lo siguiente:
\begin{eqnarray}
- V_{in} + Z_R I+ Z_C I = 0\\
I = \frac{V_{in}}{Z_R + Z_C} \\
I = \frac{V_{in}}{R + \frac{1}{j \omega C}} \\
V_0(\omega) = I Z_C \\
V_0(\omega) = \frac{V_{in}}{R + \frac{1}{j \omega C}} \frac{1}{j \omega C} \\
V_0(\omega) = \frac{j \omega CV_{in}}{Rj \omega C + 1} \frac{1}{j \omega C}\\
V_0(\omega) = \frac{j \omega CV_{in}}{R(j \omega C)^2 + j \omega C}\\
V_0(\omega) = \frac{j \omega CV_{in}}{j \omega C - R(\omega C)^2}\\
V_0(\omega) = \frac{j V_{in}}{j - R\omega C}
\end{eqnarray}

La ultima expresión se puede trabajar un poco más, por lo general se trabaja en terminos de numero complejo con. Un $Z$ se representa como $Z = a + jb$, con $a,b\in \mathbb{R}$ y $j = \sqrt{-1}$. Si se hace un arreglo algebraico en la ultima expresión tendremos lo siguiente.

\begin{eqnarray}
V_0(\omega) = \frac{-j V_{in}}{ R\omega C - j} \frac{R\omega C + j}{R\omega C + j} \\
V_0(\omega) = \frac{V_{in}-j V_{in}R\omega C}{ (R\omega C)^2 + 1}\\
V_0(\omega) = \frac{V_{in}}{ (R\omega C)^2 + 1} -j \frac{ V_{in}R\omega C}{ (R\omega C)^2 + 1}
\end{eqnarray}

Definimos la función de transferencia $H(\omega) = \frac{v_{0}(\omega)}{v_{in}(\omega)}$ y de esa forma para el ejemplo se tendrá.
\begin{eqnarray}
\frac{V_0(\omega)}{V_{in}(\omega)} = \frac{\frac{V_{in}}{ (R\omega C)^2 + 1} -j \frac{ V_{in}R\omega C}{ (R\omega C)^2 + 1}}{V_{in}}\\
H(\omega) = \frac{1}{ (R\omega C)^2 + 1} -j \frac{R\omega C}{ (R\omega C)^2 + 1}
\end{eqnarray}

## Magnitud y Fase.

Si representamos un numero complejo de valor $Z = a + bj$ en el plano complejo tenemos lo siguiente.

Se define la magnitud se determina como:
    \begin{equation}
    \lvert H(\omega)\rvert\ = \sqrt{a^2 + b^2}
    \end{equation}

La fase se define como:
    \begin{equation}
    \angle{H(\omega)} = \tan^{-1}{\left(\frac{b}{a}\right)}
    \end{equation}

Para el caso del ejemplo
\begin{eqnarray}
H(\omega) = \frac{1}{ (R\omega C)^2 + 1} -j \frac{R\omega C}{ (R\omega C)^2 + 1}\\
a = \frac{1}{ (R\omega C)^2 + 1}\\
b = -\frac{R\omega C}{ (R\omega C)^2 + 1}\\
\lvert H(\omega)\rvert\ = \sqrt{\left(\frac{1}{ (R\omega C)^2 + 1}\right)^2 + \left(-\frac{R\omega C}{ (R\omega C)^2 + 1}\right)^2} \\
\lvert H(\omega)\rvert\ = \sqrt{\left(\frac{1}{ (R\omega C)^2 + 1}\right)^2 \left[1 + \left(R \omega C\right)^2\right]} \\
\lvert H(\omega)\rvert\ = \left(\frac{1}{ (R\omega C)^2 + 1}\right)\sqrt{ 1 + \left(R \omega C\right)^2}
\end{eqnarray}

Por otro lado se tiene la fase

Para el caso del ejemplo
\begin{eqnarray}
H(\omega) = \frac{1}{ (R\omega C)^2 + 1} -j \frac{R\omega C}{ (R\omega C)^2 + 1}\\
a = \frac{1}{ (R\omega C)^2 + 1}\\
b = -\frac{R\omega C}{ (R\omega C)^2 + 1}\\
\angle{H} = \tan^{-1}\left(\frac{-\frac{R\omega C}{ (R\omega C)^2 + 1}}{\frac{1}{ (R\omega C)^2 + 1}} \right)\\
\angle{H} = \tan^{-1}\left( -R\omega C \right)
\end{eqnarray}
De lo cual se puede obtener la siguiente gráfica si $C = 100\mu F$, $R = 100\Omega$

```python
%pylab inline
import numpy as np
import pylab as pl
C = 100e-6
R = 1e2
w = np.arange(0.0001,4/(R*C),0.0001)
angulo = np.arctan(-R*w*C)
pl.figure(figsize=(16, 10), dpi=1000)
pl.xlabel('$\omega [rad/s]$', fontsize = 20)
pl.ylabel('$H$', fontsize = 20)
pl.plot(w,angulo)
pl.plot(np.array([0,4/(R*C)]),np.array([-np.pi/2 , -np.pi/2]),'--')
pl.show()
```
