Como hemos visto resolver un circuito con diodos es algo complicado. En si se necesita emplear métodos numéricos para hallar el punto $Q$, sin embargo, es posible emplear modelos que simplifican los cálculos y de ese modo encontrar soluciones aproximadas.

## Primer Modelo: Diodio Ideal

Sea $V_D$ el voltaje entre los terminales del diodo e $I_D$ la corriente que circula a través de él, entonces:
    
   \begin{equation}
     V_D = 0V \ si \ I_D>0A \ (cortocircuito)
   \end{equation}
   \begin{equation}
     I_D = 0A \ si \ V_D < 0V \ (circuito \ abierto)
   \end{equation}

```python
from __future__ import division
%pylab inline
import numpy as np
def recta(x1,y1,x2,y2):
    m = (y2-y1)/(x2-x1)
    Ts = 0.004
    x = np.arange(x1,x2,Ts)
    y = m*(x-x1) + y1
    return x, y
def constante_p(x_0,y_0,y_1):
    x = x_0*np.ones(10000)
    y = np.linspace(y_0,y_1,len(x))
    return x,y
import pylab as pl
import matplotlib.pyplot as plt
v1 = -10
v2 = 0.0
i1 = 0
i2 = 0
V0, I0 = recta(v1,i1,v2,i2)
V1,I1= constante_p(v2,0,10)
V = np.concatenate([V0,V1])
I = np.concatenate([I0,I1])
pl.plot(V,I)
ax = pl.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
pl.title('Primer modelo de corrientes vs voltaje')
pl.xlabel('$V_D$, en Volts')
pl.ylabel('$I_D$, en Amperes')
pl.figure(figsize=(10, 6), dpi=1000)
pl.rc("font", size=16)
pl.show()
```

## Segundo Modelo

Sean $V_D$ e $I_D$ el voltaje entre los terminales del diodo y la corriente que circuila a través de él, respectivamente. La relación queda definida como.

\begin{eqnarray}
I_D = 0A, \ si \  V_D<V_{\gamma} \\
V_D = V_{\gamma},  \ si \ I \ge 0
\end{eqnarray}

 En donde $V_{\gamma}$ es el voltaje de umbral para el cual el diodio puede conducir. El volor de $V_{\gamma}$ se define como:

 $\begin{equation}V_{\gamma} = \left \{ \begin{matrix} 0,3V & \mbox{si } \mbox{ es diodo de Ge}
\\ 0,7V & \mbox{si } \mbox{es diodo de Si}\end{matrix}\right. \end{equation}$

Esto es válido para diodos rectificadores, los diodos emisores de luz (LED), por ejemplo tienen otros valores de umbral

```python
# Diodo de Silicio
v1si = -5
v2si = 0.7
i1si = 0
i2si = 0
V0si, I0si = recta(v1si,i1si,v2si,i2si)
V1si,I1si= constante_p(v2si,0,10)
Vsi = np.concatenate([V0si,V1si])
Isi = np.concatenate([I0si,I1si])
# Diodo de Germanio
v1ge = -5
v2ge = 0.3
i1ge = 0
i2ge = 0
V0ge, I0ge = recta(v1ge,i1ge,v2ge,i2ge)
V1ge,I1ge= constante_p(v2ge,0,10)
Vge = np.concatenate([V0ge,V1ge])
Ige = np.concatenate([I0ge,I1ge])

pl.plot(Vsi,Isi,Vge,Ige)
ax = pl.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
pl.title('Segundo modelo de corrientes vs voltaje en diodo')
pl.xlabel('$V_D$, en Volts')
pl.ylabel('$I_D$, en Amperes')
si = '$D_{Si}$'
pl.legend(si)
pl.legend(['$I$ en Diodo de $Si$','$I$ en Diodo de $Ge$' ], loc=2)
#pl.figure(figsize=(10, 6), dpi=1000)
pl.rc("font", size=15)
pl.show()
```

## Tercer modelo de diodo

Consideremos que la corriente del diodo es $I = I_0 \left( e^{\frac{V}{\eta V_T}} - 1 \right)$ Graficamente para un diodo de silicio tenemos:

Por ley de Ohm tenemos:

$\begin{equation} 
I = \frac{V}{R} \Leftrightarrow R = \frac{V}{I}
\end{equation}$

Definimos la resisencia en corriente continua del diodo como: Sea $V$ el voltaje entre los terminales del diodo, e $I$ la corriente que circula a través de él. Entonces:

$\begin{equation} 
R_{CD} = \frac{V}{I}
\end{equation}$

Ejemplo: Consideremos el siguiente gráfico, determine la resistencia en CD sí $V = 0,95V$ para un diodo de Silicio a temperatura ambiente $(300k)$ con $I_0 = 4nA$

```python
%pylab inline
import numpy as np
import pylab as pl
n = 2
VT = 26e-3
V = np.arange(0.65,1,0.00001)
I0 = 4e-9
I = I0*(np.exp(V/(n*VT))-1)
pl.figure(figsize=(15, 6), dpi=1000)

pl.plot(V,1000*I)

VD = 0.95 # abscisa 
ID = 1000*I0*(np.exp(VD/(n*VT))-1)




pl.scatter([VD, ],[ID, ], 50, color='red')
pl.annotate(r'$Q\left(0,95V \ ; \ 4 \times 10^{-6} \times \left(e^{0.95V/(2 \times 26 mV)}-1\right) mA \right)$',
            xy=(VD, ID), xycoords='data',
            xytext=(-250, +120), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.35"))



pl.rc("font", size=15)
pl.plot(np.array([VD,VD]), np.array([0,ID]), color='blue', linewidth=.5, linestyle="--")



ax = pl.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0.65))
pl.ylabel('$I$, en $mA$')
pl.xlabel('$V$, en $V$')
pl.show()
```
